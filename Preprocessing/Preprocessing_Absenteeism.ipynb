```python
#Importing libraries
import pandas as pd
```

```python
#Loading the CSV file
df = pd.read_csv('Absenteeism_data.csv')

df.head()
```

```python
#Dropping the id column
del df['ID']

df.head()
```

```python
#Turning the reasons for absence to dummy variables in groups of...
#1-14 ,15-17, 18-21 and 22-28
#First replacing those groups all into an individual number using replace
groups=[list(range(1,15)),list(range(15,18)),list(range(18,22)),list(range(22,29))]    

for i in range(4):
    for item in groups[i]:
        df['Reason for Absence'].replace(item,i+1,inplace=True)

#Create dummy variables and add to the table
dummy = pd.get_dummies(df,prefix = "Reason", prefix_sep="_", columns = ['Reason for Absence'])
df = pd.concat([df,dummy],axis= 1)
df.head()
```

```python
#Drop reason for absence column and reason 0 because it's just people with no reasons for absence
df.drop(['Reason for Absence', 'Reason_0'],axis = 1,inplace=True)
df.head()
```

```python
df.drop_duplicates(keep ='first')

#Extract the date and month vlues then drop the date column

dates = df['Date'].values.tolist()#Creates list with each element shown twice
dateslist =[]
[dateslist.append(dates[i][0]) for i in range(len(dates))]#Remove duplicates
print(dateslist)
```

```python
#Converted series to list to make applying day of the week function easier
import datetime
month= []
day = []
for a in range(len(dateslist)):
    month.append(dateslist[a][3:5])
    day.append(datetime.date(int(dateslist[a][6:10]),int(dateslist[a][3:5]),int(dateslist[a][0:2])).weekday())
```

```python
#Adding day of week and month to dataframe and dropping date column
df['Day of the week'] = pd.DataFrame(day)
df['Month'] = pd.DataFrame(month)
df.drop(['Date'],axis = 1, inplace = True)
```

```python
df = df.loc[:,~df.columns.duplicated()]
df
```

```python
#Changing education 1 to 0 and everything else to 1
df['Education'].replace(1,0,inplace=True)
[df['Education'].replace(b,1,inplace=True) for b in range(2,5)]
df
```

```python
#Rearranging the columns
cols = df.columns.tolist()
cols
```

```python
cols = cols[9:13]+cols[:-3:-1]+cols[0:9]
cols
```

```python
df = df[cols]
df
```

```python
#Changing all to numeric values
df.apply(pd.to_numeric).info()
```

```python
#Saving the csv file
df.to_csv('preprocssed_absenteeism.csv',index=False)
```
